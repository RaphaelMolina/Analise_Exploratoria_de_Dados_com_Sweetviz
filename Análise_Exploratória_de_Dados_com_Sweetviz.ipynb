```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Gerar dados aleatórios e plotar histagrama (distribuição)

```python
minutos_por_dia = np.random.randint(0, 60, 30)
```

```python
minutos_por_dia
```

```python
aux = sns.histplot(minutos_por_dia)
aux.set(xlabel='Minutos por dia', ylabel='Quant. de vezes')
plt.show()
```

# Carregar um arquivo de base dos jogos de tenis de 2021

https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv

```python
arquivo = 'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv'
```

```python
dados = pd.read_csv(arquivo)
dados
```

```python
analise = sweetviz.analyze(dados)
analise.show_html()
```

# Gerar uma matriz de correlação (de Pearson) de variáveis

```python
mascara = np.triu(np.ones_like(dados.corr(), dtype=np.bool))
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dados.corr(), mask=mascara, square=True, annot=True, vmin=-1, vmax=1)
plt.show()
```
